### Seprate the input and output variables

##### three inputs & One Output

```python
x = df[['disp','hp','wt']]              #displacement , horsepower, weight
y = df['mpg']
```

```python
x
```

### import the class

```python
from sklearn.linear_model import LinearRegression
```

#### create the object

```python
reg = LinearRegression()
```

#### train the model

```python
reg.fit(x,y)
```

#### predict on unseen data

```python
disp = 221
hp = 102
wt = 3.51        #inversly propotion to mileage - wt increase mileage decrease & vice versa
reg.predict([[disp, hp, wt]])
```

#### visualize

```python
plt.figure(figsize=(16,9))
plt.subplot(2,2,1)
plt.scatter(df['disp'], y, color='r')
plt.subplot(2,2,2)
plt.scatter(df['hp'], y, color='b')
plt.subplot(2,2,3)
plt.scatter(df['wt'], y, color='g')
```

```python
reg.score(x, y)*100           # Accuracy Score is 82%
```

### Decision Tree Classification

#### import the dataset

```python
df = pd.read_csv('Social_Network_Ads.csv')
```

```python
df.shape
```

```python
list(df.columns)
```

```python
df           # 0 means >>> Purchased,  1 means>>> Not Purchased......
```

#### Seprate the input & Output Variables

```python
x = df[['Age',"EstimatedSalary"]]
y = df['Purchased']
```

```python
x
```

#### Explore the data

```python
plt.xlabel('Age')
plt.ylabel('Salary')
plt.grid()
plt.scatter(x['Age'], x['EstimatedSalary'], color='r')
```

### import the class

```python
from sklearn.tree import DecisionTreeClassifier
```

#### create the object

```python
classifier = DecisionTreeClassifier(random_state=0)
```

#### train the model

```python
classifier.fit(x, y)
```

#### Plot the tree

```python
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(15,15))
plot_tree(classifier, fontsize=7, feature_names=['age','sal'],
         class_names=['y','n'], filled=True, rounded=True);
```

#### Prediction on unseen data

```python
new = pd.DataFrame({
    'Age': [25,37,56,66],
    'EstimatedSalary': [34000,41000,26000,134000]
})
```

```python
new
```

```python
classifier.predict(new)
```
